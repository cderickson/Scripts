```python
import pickle
import pandas as pd

with open('ALL_DECKS', 'rb') as f:
    data = pickle.load(f)

# See what type the pickle loaded as
print(type(data))
```

```python
def pprint_dict_structure(obj, indent=0, skip_leaves=True):
    """Pretty print dict/list structure; summarize lowest-level collections (e.g. set of card names)."""
    prefix = "  " * indent
    if isinstance(obj, dict):
        for k, v in obj.items():
            if skip_leaves and isinstance(v, (list, set)) and v:
                sample = next(iter(v)) if isinstance(v, set) else v[0]
                if isinstance(sample, (str, int, float, bool)):
                    print(f"{prefix}{k!r}: <{type(v).__name__} of {len(v)} items>")
                    continue
            print(f"{prefix}{k!r}:")
            pprint_dict_structure(v, indent + 1, skip_leaves)
    elif isinstance(obj, list):
        for i, item in enumerate(obj):
            if skip_leaves and isinstance(item, (list, set)) and item:
                sample = next(iter(item)) if isinstance(item, set) else item[0]
                if isinstance(sample, (str, int, float, bool)):
                    print(f"{prefix}[{i}]: <{type(item).__name__} of {len(item)} items>")
                    continue
            print(f"{prefix}[{i}]:")
            pprint_dict_structure(item, indent + 1, skip_leaves)
    else:
        print(f"{prefix}{obj!r}")


pprint_dict_structure(data)
```
